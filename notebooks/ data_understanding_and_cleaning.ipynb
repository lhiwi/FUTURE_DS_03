## Create a feature for previous contact

The `pdays` variable uses -1 to indicate that the customer was not contacted in a previous campaign.  
We create a new binary feature to make this easier to interpret in analysis.

```python
df['was_contacted_before'] = df['pdays'].apply(lambda x: 'no' if x == -1 else 'yes')
```

```python
df['was_contacted_before'].value_counts()
```

## Create conversion flag

The target column `y` contains yes/no values. We create a numeric conversion flag where converted customers are coded as 1 and non-converted customers are coded as 0. This makes KPI calculations easier later.

```python
df['conversion_flag'] = df['y'].map({'yes': 1, 'no': 0})
```

```python
df[['y', 'conversion_flag']].head()
```
